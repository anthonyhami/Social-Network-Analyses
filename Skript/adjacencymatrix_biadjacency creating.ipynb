#  5 full

```python
df=pd.read_csv("choosen_disease_5_6_most_final.csv")
```

```python
t1 = pd.crosstab(df['Disease_ID'], df['Gene_ID'])
t2 = pd.crosstab(df['Gene_ID'], df['Disease_ID'])

df1 = t1.combine_first(t2)
```

```python
df1
```

```python
cols = [c for c in df1.columns if  isinstance(c, int) ]
df1=df1[cols]
df1=df1.dropna()
df1
```

sns.heatmap(df1, cmap ='gray', linewidths = 0.30, annot = True)

```python
sns.set(rc={'figure.figsize':(12,8)})
fig=sns.heatmap(df1, cmap ='gray', linewidths = 0.30, annot = True)
fig.figure.savefig("output_all.png")
```
