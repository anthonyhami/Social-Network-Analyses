# Create the Disease network for DF 3 (5 Gene) for cytoscape

```python
df_3
```

```python
df_disease_network=pd.DataFrame(columns=["Source_Disease","Target_Disease"])
grouped = df_3.groupby('Disease_ID') #Gene_ID
for name, group in grouped:
    #print(group)
    print(name)
    related_gene=group["Gene_ID"].to_list() #Disease_ID
    for elem in related_gene:
        related_disease=df_3[df_3["Gene_ID"]== elem]["Disease_ID"].to_list() #tauschen
        print(related_disease)
        
        
        for elem_disease in related_disease:
            #A ->B but not B-> or B->A but not A->B
            probe=df_disease_network[(df_disease_network['Source_Disease']==elem_disease)&
                                     (df_disease_network['Target_Disease'] == name)]
            print(probe)
            if probe.empty and (name!=elem_disease): # if (name!=elem_disease): bei plot
                
                df_disease_network.loc[len(df_disease_network)]=[name, elem_disease]
                
df_disease_network=df_disease_network.drop_duplicates()
        
        
        
    
    
    


#skip_Disease=[]
#for index,row in df_3.iterrows():
    #Disease,Gene= row["Disease_ID"],row["Gene_ID"]
    
    
```
