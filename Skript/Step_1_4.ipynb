# Filter all Diseases that have a high degree(100 genes or more) but have the genes that occur in the unique_six_two data

```python
skip_Disease=[]
for index,row in unique_six_two.iterrows():
    Disease,Gene= row["Disease_ID"],row["Gene_ID"]
    #print(Disease)
    #print(Gene)
    #schauen für jede Gene mit wievielen Disease es verbunden ist
    unique_six_two_Gene_filter=unique_six_two[unique_six_two["Gene_ID"]==Gene] #nach Genen Gruppieren
    print(unique_six_two_Gene_filter)
    for index_2,row_2 in unique_six_two_Gene_filter.iterrows():
        Disease_2= row_2["Disease_ID"]
        unique_six_two_Gene_Disease_filter=df[df["Disease_ID"]==Disease_2]
        print(Disease_2,len(unique_six_two_Gene_Disease_filter))
        if len(unique_six_two_Gene_Disease_filter)<100:  #alle Diseases die weniger als 100 gene haben werden rausgeschmießen
            skip_Disease.append(Disease_2)
    print()
```

```python
len(set(skip_Disease))
#unique disease anstatt doppelung
#467 with <100
#soviele Diseases rausgeworfen
```

# 467 Diseases have been droped out

```python

unique_six_two_hub=unique_six_two[~unique_six_two['Disease_ID'].isin(list(set(skip_Disease)))]
```
